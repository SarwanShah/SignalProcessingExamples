```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
pi = 3.141592654

def myExp(x):
  approxOrder = 15
  factorialTerm = 1
  powerTerm = x
  result = 1 + x

  for i in range(2, approxOrder):
    powerTerm = x * powerTerm
    factorialTerm = i * factorialTerm
    result += powerTerm / factorialTerm

  return result

def mySin(x):
  x = normalizeRad(x)
  approxOrder = 14
  positiveTerm = 0
  factorialTerm = 1
  powerTerm = x
  result = x

  for i in range(3, approxOrder, 2):
    if (positiveTerm):
      powerTerm = x * x * powerTerm;
      factorialTerm = i * (i - 1) * factorialTerm;
      result += (powerTerm / factorialTerm);
      positiveTerm = 0;
    else:
      powerTerm = x * x * powerTerm;
      factorialTerm = i * (i - 1) * factorialTerm;
      result -= (powerTerm / factorialTerm);
      positiveTerm = 1;
    

  return result;

def myCos(x):
  x = normalizeRad(x)
  approxOrder = 14
  positiveTerm = 0
  factorialTerm = 1
  powerTerm = x
  result = 1

  for i in range(2, approxOrder, 2):
    if (positiveTerm):
      powerTerm = x * x * powerTerm;
      factorialTerm = i * (i - 1) * factorialTerm;
      result += (powerTerm / factorialTerm);
      positiveTerm = 0;
    else:
      powerTerm = x * x * powerTerm;
      factorialTerm = i * (i - 1) * factorialTerm;
      result -= (powerTerm / factorialTerm);
      positiveTerm = 1;
    

  return result;

def normalizeRad(x):
  if (x > pi):
    while (x > pi):
      x -= 2 * pi
    
  elif (x < -pi):
    while (x < -pi):
      x += 2 * pi


  return x
```

```python
N = 1000
x = sorted(np.random.uniform(-np.pi, np.pi, N))
ySinApprox = []
yCosApprox = []
ySinActual = []
yCosActual = []

for i in range(N):
    ySinApprox.append(mySin(x[i]))
    ySinActual.append(math.sin(x[i]))
    yCosApprox.append(myCos(x[i]))
    yCosActual.append(math.cos(x[i]))
        
plt.plot(x, ySinApprox, color='b', linestyle='-', label='Approximation')
plt.plot(x, ySinActual, color='y', linestyle='--', label='Actual')
plt.title("Comparison of 15th order Taylor Series Approximation of Sin with Actual Function")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.figure()
plt.plot(x, abs(np.array(ySinActual) - np.array(ySinApprox)), label="Error")
plt.title("Error in Actual vs Approximate Sin (14th Order Approx)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()


    
    
```

```python
N = 1000
x = sorted(np.random.uniform(0, 12, N))
yExpApprox = []
yExpActual = []

for i in range(N):
    yExpApprox.append(myExp(x[i]))
    yExpActual.append(math.exp(x[i]))

        
plt.plot(x, yExpApprox, color='b', linestyle='-', label='Approximation')
plt.plot(x, yExpActual, color='y', linestyle='--', label='Actual')
plt.title("Comparison of 10th order Taylor Series Approximation of Exp with Actual Function")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.figure()
plt.plot(x, abs(np.array(yExpActual) - np.array(yExpApprox)), label="Error")
plt.title("Error in Actual vs Approximate Exp")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()


    
    
```
